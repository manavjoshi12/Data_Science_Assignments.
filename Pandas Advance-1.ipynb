# Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The function should print the sum to the console.

```python
def sum_first_three(df1):
    values = df1['Values'].values
    sum_first_three = sum(values[:3])
    print(f"The sum of the first three values in the 'Values' column is {sum_first_three}")
```

```python
df1 = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
sum_first_three(df1)
```

# Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column 'Word_Count' that contains the number of words in each row of the 'Text' column.

```python
def add_word_count(df):
    word_count_fn = lambda text: len(text.split())
    df['Word_Count'] = df['Text'].apply(word_count_fn)
    
    return df

    
```

```python
df = pd.DataFrame({'Text': ['This is a sentence.', 'Another sentence.', 'A third sentence.']})

df = add_word_count(df)

print(df)

```

# Q5. How are DataFrame.size() and DataFrame.shape() different?

DataFrame.size() and DataFrame.shape() are both methods used in Pandas to retrieve information about the dimensions of a DataFrame, but they return different types of information.

* DataFrame.size() returns the total number of elements in the DataFrame, which is equal to the number of rows multiplied by the number of columns. This method returns a single integer value.

* DataFrame.shape() returns a tuple of integers representing the number of rows and columns in the DataFrame, respectively. For example, if a DataFrame has 10 rows and 5 columns, DataFrame.shape() would return (10, 5).

In summary, DataFrame.size() returns the total number of elements in the DataFrame, while DataFrame.shape() returns a tuple of the number of rows and columns in the DataFrame.

# Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address. The username is the part of the email address that appears before the '@' symbol. For example, if the email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your function should extract the username from each email address and store it in the new 'Username' column.

```python
def extract_username(df4):
    df4['Username'] = df4['Email'].apply(lambda x: x.split('@')[0])
    return df4
```

```python
data4 = {"Email" : ['john.doe@example.com'
'jane.smith@hotmail.com',
'bob.johnson@gmail.com',
'sarah.miller@yahoo.com',
'alex.chan@outlook.com']}
df4 = pd.DataFrame(data4)
df4 = extract_username(df4)
print(df4)
```

# Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

```python

df7 = pd.DataFrame({
    'Date': pd.date_range(start='2022-02-01', periods=20, freq='D'),
    'Sales': [10, 15, 25, 30, 20, 25, 35, 45, 55, 60, 65, 70, 80, 75, 85, 90, 80, 70, 60, 50]
})


df7['MovingAverage'] = df7['Sales'].rolling(window=7, min_periods=1).mean()

print(df7)
```

# Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g. Monday, Tuesday) corresponding to each date in the 'Date' column.


```python
'''
For example, if df contains the following values:

    Date
0 2023-01-01
1 2023-01-02
2 2023-01-03
3 2023-01-04
4 2023-01-05

Your function should create the following DataFrame:

Date Weekday
0 2023-01-01 Sunday
1 2023-01-02 Monday
2 2023-01-03 Tuesday
3 2023-01-04 Wednesday
4 2023-01-05 Thursday
The function should return the modified DataFrame.
'''
```

```python
data8 = {"Date" : ['2023-01-01',
 '023-01-02',
 '2023-01-03',
 '2023-01-04',
 '2023-01-05']
 }
```

```python
df8 = pd.DataFrame(data8)
```

```python
df8['Date'] = pd.to_datetime(df8['Date'], errors='coerce')
df8 = df8.dropna(subset=['Date'])
```

```python
df8['Weekday'] = df8['Date'].dt.day_name()
```
