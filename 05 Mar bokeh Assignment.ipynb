# Q1. How can you create a Bokeh plot using Python code?

# To create a Bokeh plot using Python code, we need to follow the following steps:

* Install the Bokeh library using pip install bokeh.

* Import the necessary modules such as figure from bokeh.plotting, and output_file and show from bokeh.io.

* Create a data source for the plot using bokeh.sampledata.iris import flowers(iris = name of the dataset).

* Create a figure using the figure() function and specifying the plot attributes.

* Add the glyphs to the figure using the figure() function and specifying the data source and plot attributes.

* Save the output to an HTML file using the output_file() function.
Display the output using the show() function.


Here is an example code to create a simple Bokeh plot:

```python
! pip install bokeh
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure, output_file , show
from bokeh.sampledata.iris import flowers
```

```python
p = figure()
p.circle(flowers['petal_length'] , flowers['petal_width'])
output_file('text.html')
show(p)
```

# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

In Bokeh, glyphs are visual shapes that can be added to a plot to represent data points. The most common glyphs include circles, squares, triangles, and lines. To add glyphs to a Bokeh plot, we first need to create a data source using the ColumnDataSource class. Then we can call the appropriate glyph function (e.g., circle(), square(), etc.) and pass the data source along with the column names for the x and y coordinates.

Here is an example:

```python
p = figure()
p.triangle(flowers['petal_length'] , flowers['petal_width'])
output_file('text.html')
show(p)
```

# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

In Bokeh, you can customize the appearance of a plot using various properties of the plot, axis, title, and legend. Here are some examples:

Axis Properties: You can customize the labels, tick marks, and grid lines of the x and y axis using properties like axis_label, major_label_text_color, major_tick_line_color, minor_tick_line_color, axis_line_color, axis_line_width, etc.

Title Properties: You can customize the font, color, and position of the plot title using properties like title, title_text_font_size, title_text_color, title_location, etc.

Legend Properties: You can customize the font, color, and position of the legend using properties like legend, legend_location, legend_label_text_font_size, legend_label_text_color, etc. 

These are just a few examples of how you can customize the appearance of a Bokeh plot. Bokeh provides a wide range of properties and tools for creating beautiful and interactive visualizations.

# Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time ?

Bokeh server is a way to create and serve interactive Bokeh plots that can be updated in real-time. The Bokeh server allows you to add interactivity to your plots by providing a way to connect to a Python function or data source that can modify or update the plot's data or properties.

To use the Bokeh server, you need to create a Python script that defines the plot, the data, and the interactive features you want to include. Then, you can run the script using the Bokeh server command-line tool or integrate the Bokeh server into a larger web application.

Once the Bokeh server is running, you can view the plot in a web browser and interact with it by clicking, hovering, or selecting data points. Any updates or changes made to the plot will be automatically reflected in the browser in real-time.

# Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

To embed a Bokeh plot into a web page or dashboard using Flask or Django, you need to follow these general steps:

Create a Bokeh plot using Python code and save it to an HTML file using the output_file function in Bokeh.
In your Flask or Django app, define a view function that will serve the HTML file with the Bokeh plot.
In the view function, use the render_template function in Flask or the render function in Django to render the HTML file and pass any necessary variables.
Create a route in Flask or a URL pattern in Django that will map to the view function.
In your HTML template file, use the {{ url_for }} function in Flask or the url function in Django to link to the route or URL pattern for the view function.
In the HTML template file, use an iframe to embed the HTML file with the Bokeh plot.
