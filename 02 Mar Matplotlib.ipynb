# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a data visualization library in Python that provides a wide variety of customizable, high-quality 2D and 3D visualizations for generating static, animated, and interactive visualizations. It is a powerful tool for creating plots, histograms, power spectra, bar charts, error charts, scatterplots, and many other types of visualizations.

Matplotlib is widely used in the scientific community, data science, and engineering fields for creating visualizations that are easily interpretable and communicable to a broad audience. It is also used in industry for generating visualizations for reports, presentations, and dashboards.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

* Line Plot: A basic plot used for showing the trend in a data set over time or another continuous variable.

* Scatter Plot: A plot used for showing the relationship between two continuous variables.

* Bar Plot: A plot used for showing comparisons between discrete categories.

* Histogram: A plot used for showing the distribution of a continuous variable.

* Pie Chart: A plot used for showing the composition of a whole in terms of its parts.

# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


A scatter plot is a type of graph that displays the relationship between two variables by plotting them as points on a two-dimensional coordinate system, where each point represents a single observation. The horizontal axis represents the independent variable, while the vertical axis represents the dependent variable. Scatter plots are commonly used to identify patterns or relationships between variables, such as whether there is a correlation or association between them. The patterns that emerge from the data points on the scatter plot can help to identify trends, outliers, clusters, or gaps in the data. Scatter plots can be used in various fields, such as statistics, data analysis, economics, social sciences, and engineering, to name a few.

```python
import numpy as np
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y , c="#8A6FA0" , alpha= .8)
plt.xlabel("This is my x-axis")
plt.ylabel("This is my y-axis")
plt.title("This is my scatter plot")
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

The subplot() function is used to create multiple plots in the same figure. It allows us to divide the figure into multiple subplots and plot different graphs on each subplot.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots
fig, axs = plt.subplots(2, 2)

# Plot the first line
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot the second line
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot the third line
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot the fourth line
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a title to the figure
fig.suptitle('Four Line Plots')

# Display the plots
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


A bar plot is a type of plot used to visualize categorical data with rectangular bars with lengths proportional to the values that they represent. The bars can be oriented vertically or horizontally. A bar plot can be used to display the frequency or count of categories in a dataset, as well as compare the values of different categories.

Bar plots are commonly used in data visualization for the following reasons:

* They are simple and easy to understand, even for non-technical audiences.
* They are effective at displaying relative differences between groups or categories.
* They can be used with both categorical and numerical data.
* They can be customized to show additional information, such as error bars, confidence intervals, or annotations.
* Some common variations of bar plots include grouped bar plots, stacked bar plots, and horizontal bar plots.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

# Bar Plot

```python
plt.figure(figsize=(6,2))
plt.bar(company, profit , color = "#6F89A0")
plt.xlabel('Company')
plt.ylabel('Profit')
```

# Horizontal Bar Plot

```python
plt.figure(figsize=(6,2))
plt.barh(company, profit , color = "#A06F98")
plt.xlabel('Profit')
plt.ylabel('Company')
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset that shows the distribution of the data using quartiles. It is used to visually summarize the distribution of a dataset and to identify any outliers.

The box plot represents five statistical summary values of the dataset: the median (the value that separates the upper 50% and the lower 50% of the data), the upper quartile (Q3, the value that separates the upper 25% of the data from the lower 75%), the lower quartile (Q1, the value that separates the lower 25% of the data from the upper 75%), the minimum value (the lowest data point within 1.5 times the interquartile range (IQR) of the lower quartile), and the maximum value (the highest data point within 1.5 times the IQR of the upper quartile).

Box plots are useful for detecting outliers and comparing the distribution of data across different categories or groups. They are commonly used in statistical analysis, data science, and data visualization.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data1 = [box1, box2]

fig, ax = plt.subplots()

ax.boxplot(data1)

plt.show()
```
